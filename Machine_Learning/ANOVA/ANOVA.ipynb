ANOVA visual

```python
from pandas import read_csv
import os
import pandas as pd

#import for visualization
import seaborn as sns # for plotting graphs
import matplotlib
import matplotlib.pyplot as plt # for plotting graphs
%matplotlib inline
```

```python
## Read the Dataset

df=pd.read_csv('anova_1.csv')
df.head()
```

```python

A = df["ClassA"]
B = df["ClassB"]
C = df["ClassC"]

label1 = "Class A"
label2 = "Class B"
label3 = "Class C"
title = 'ANOVA Distribution'
#-------

# Plot Distribution 
plt.figure(figsize=(10,5))
sns.distplot(A,hist=False, bins=10, color='g', label = label1).set_title(title)
sns.distplot(B,hist=False, bins=10, color='b', label = label2).set_title(title)
sns.distplot(C,hist=False, bins=10, color='r', label = label3).set_title(title)
```

```python
## Read the Dataset

wses_df=pd.read_csv('WSES_data.csv')
wses_df.head()
```

```python
# Fetch Ratings for Big/Small Production movies
wses_in = wses_df[(wses_df["Region"] == 'India')]
wses_na = wses_df[(wses_df["Region"] == 'Americas')]
wses_uk = wses_df[(wses_df["Region"] == 'UK')]

india = wses_in["Profit %"]
americas = wses_na["Profit %"]
uk = wses_uk["Profit %"]


# Plot Distribution 
plt.figure(figsize=(10,5))
sns.distplot(india,hist=False, bins=10, color='g', label = "India").set_title(title)
sns.distplot(americas,hist=False, bins=10, color='b', label = "Americas").set_title(title)
sns.distplot(uk,hist=False, bins=10, color='r', label = "UK").set_title(title)
```
